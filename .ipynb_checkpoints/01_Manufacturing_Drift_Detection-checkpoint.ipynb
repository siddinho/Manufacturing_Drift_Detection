# Machine Tester Drifter Problem

The machine/tester drift problem
Modern manufacturing plants employ enormously complex machines and equipment for production, packaging, testing, etc. Although designed for high quality and stability, no machine is immune to drift and variance. Over time, sub-components start to behave slightly differently than what they did in the installation phase, and the overall characteristic drifts.



Detecting this drift is important for many reasons,
to identify which machine (among a set of them) may require repair and maintenance
to correct/reset the drift by software or hardware intervention
to correlate product quality with the machine drift
to investigate yield issues or reliability failures that might have been caused by the machine drift and increased variance

## Individual drift detection
Detecting drift and change for an individual piece of equipment is a common manufacturing problem. This can be handled using various approaches, especially if some sort of test/sensor data is also available. A few approaches are,<br>
1. looking for and counting anomalies in the process — if their density exceeds a threshold, then the machine might need repair/maintenance <br>
2. monitoring the statistical distribution (not necessarily Normal/Gaussian) of the data associated with a machine and detecting when that changes <br>
3. performing some sort of simple statistical tests (e.g. comparing means to a golden standard machine) every day/week and detecting if there is a significant change

**Which machines are falling out most? **<br>
Often, the most prudent task is to identify the machines (among a set of many identical machines) that are drifting most compared to a golden standard. Ideally, we want to rank them by assigning some sort of numerical score.<br>
This is often practiced in real-life scenarios. Why? Because previous experience and data analytics might have shown that a small drift does not impact the product quality or yield, and only the machines which have drifted most should be stopped for intervention and maintenance.<br>
Business managers and manufacturing VPs will tell you that intervention costs money and should only be done when unavoidable. Your data analytics, they expect, will guide them on the most pressing problems, and not just throw general statistical analyses and charts at them.

Let’s say, <br>
we collect 200 samples’ measurements per machine <br>
there are 10 process measurements (or sensors) — A, B, C, etc.<br>
there are 10 machines — M1, M2, M3, etc.<br>
there is, of course, a golden machine, with which we want to benchmark all these 10 machines and detect drift <br>
So, we have a nice table of 20,000 data points (plus the golden machine data). But each measurement is different in nature and does not correlate to each other. The measurement from sensor ‘A’ in machine ‘M1’ correlates to the measurement of sensor ‘A’ from machine ‘M2’, or that from the machine ‘M3’.

 It is not a clean-cut ML problem where you have a labeled dataset of “machine drifts”. Here, you have to carefully construct statistical tests and measures to detect drift and rank machines based on that measurement.

Again, like any other data-driven problem-solving case, you have options. <br>
1. you can extract descriptive statistics from each sensor data column and pairwise compare them to those from the other machines
2. you can calculate some sort of global ‘distance score’ for each machine from the golden machine (here is a list of distance metrics that you can evaluate using the Scipy package)
3. you can compute sophisticated distance metrics between individual data distributions (e.g., Mahalanobis distance) and rank based on them

# Code Changed Normal Distribution to Random on own

```python
import pandas as pd
import numpy as np, matplotlib.pyplot as plt
```

```python
machine_golden = pd.DataFrame(np.random.normal(size = (200,10)))
```

### Golden Machine Data

```python
machine_golden.head()
```

# Generating DataFrame of 10 different machines 

```python
machines = {'machine'+str(i):[] for i in range(1,11)}
for i in range(1,11):
    loc = np.random.uniform(0,2)
    scale = np.random.uniform(0,2)
    df1 = machine_golden+pd.DataFrame(np.random.normal(loc,scale,size=(200,10)))
    #df1 = machine_golden + pd.DataFrame(np.random.rand(200,10))
    machines['machine'+str(i)] = df1
```

```python
df1.shape
```

```python
df1.head()
```

# Visualizing Sensor 1 of Machine M1 and Golden Machine Sensor M1

```python
plt.figure(figsize=(5,3),dpi = 120)
plt.scatter(machine_golden.loc[:,0],machines['machine1'].loc[:,0], 
            edgecolor='k', color='#42a1f5')
plt.xlabel("Sensor 1 of Golden Machine ")
plt.ylabel("Sensor 1 of Machine M1 ")
plt.title("Sensor-1 of golden machine vs. Sensor-1 of $M$1")
plt.show()
```

```python
plt.figure(figsize=(5,3),dpi=120)
plt.scatter(machine_golden.loc[:,0],machines['machine1'].loc[:,1], 
            edgecolor='k', color='red')
plt.xlabel("Sensor 1 of Golden Machine ")
plt.ylabel("Sensor 2 of Machine M1 ")
plt.title("Sensor-1 of golden machine vs. Sensor-2 of $M$1")
plt.show()
```

```python
plt.figure(figsize=(10,3),dpi=120)
plt.subplot(1,2,1)
plt.scatter(machine_golden.loc[:,0],machines['machine1'].loc[:,0], 
            edgecolor='k', color='#42a1f5')
plt.xlabel("Sensor 1 of Golden Machine ")
plt.ylabel("Sensor 1 of Machine M1 ")
plt.title("Sensor-1 of golden machine vs. Sensor-1 of $M$1")


plt.subplot(1,2,2)
plt.scatter(machine_golden.loc[:,0],machines['machine1'].loc[:,1], 
            edgecolor='k', color='red')
plt.xlabel("Sensor 1 of Golden Machine ")
plt.ylabel("Sensor 2 of Machine M1 ")
plt.title("Sensor-1 of golden machine vs. Sensor-2 of $M$1")
plt.tight_layout()
plt.show()
```

## Corelation Analysis - DataFrame to DataFrame - .corrwith()

```python
machine_golden.corrwith(machines['machine1'],axis=1) # returns pairwise correlations from 2 dataframes
```

## Getting Co-relations with all Machines

```python
for i in machines.keys():
    print("Average Co-relation of Machine {0} with Golden State Machine {1} ".format(i,
    round(machine_golden.corrwith(machines[i],axis =1).mean(),3)))
    
```

* <b> Machine 8 has the least corelation and Machine 5 has the maximum corelation <b>

# Visual Inspection

```python
fig,ax  = plt.subplots(1,10,figsize=(30,4),dpi = 120)
plt.suptitle("Golden machine and machine 8 - Least Co-related", fontsize=18,va='top')
ax = ax.ravel()
for i in range(10):
    ax[i].scatter(machine_golden.loc[:,i],machines['machine8'].loc[:,i])
    ax[i].set_xlabel(f"Sensor {i+1}",fontsize=15)
    
fig,ax  = plt.subplots(1,10,figsize=(30,4),dpi = 120)
plt.suptitle("Golden machine and machine 5 Most Co-related", fontsize=18,va='top')
ax = ax.ravel()
for i in range(10):
    ax[i].scatter(machine_golden.loc[:,i],machines['machine5'].loc[:,i])
    ax[i].set_xlabel(f"Sensor {i+1}",fontsize=15)

plt.show()
```
